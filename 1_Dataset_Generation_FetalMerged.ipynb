```python
import json
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

CONFIG = {
    "database_path": "../Fetal",
    "output_path": "../Dataset/Merged/Landmarks_3_10",
    "scale_factor": 0.10, # 10% of the original average contour size
    "resolutions": ["full", "half", "quarter"], # Names are only for reference
    # "resolutions": [1, 1/2, 1/4], # Actual scale factors
    "image_types": [""],  
    "organs": ["1", "2"], # ID in the mask
    "organ_names": ["Pubic symphysis", "Fetal head"], # Name ID
    "inputsize": 512
}

os.makedirs(CONFIG["output_path"], exist_ok=True)
with open(os.path.join(CONFIG["output_path"], "config.json"), "w") as f:
    json.dump(CONFIG, f, indent=4)

# now copy from the other datasets
data_paths = ["/home/ngaggion/SSD/Mask2Graph/Dataset/HC18/Landmarks_3_10",
              "/home/ngaggion/SSD/Mask2Graph/Dataset/JNU-IFM/Landmarks_3_10",
              "/home/ngaggion/SSD/Mask2Graph/Dataset/PSFHS/Landmarks_3_10"]

# should copy the images, landmarks, masks folders
# the read the image_list.txt, mask_list.txt, train.txt, val.txt and test.txt and merged them together

for data_path in data_paths:
    # copy the entire folders
    for folder in ["images", "landmarks", "masks"]:
        src = os.path.join(data_path, folder)
        dst = os.path.join(CONFIG["output_path"], folder)
        if os.path.exists(src):
            print(f"Copying {src} to {dst}")
            os.system(f"rsync -avzP {src}/ {dst}/")
        else:
            print(f"{src} does not exist")
    
    # now copy the image_list.txt, mask_list.txt, train.txt, val.txt and test.txt
    # but create them and concatenate them
    for file in ["image_list.txt", "mask_list.txt", "train.txt", "val.txt", "test.txt"]:
        src = os.path.join(data_path, file)
        dst = os.path.join(CONFIG["output_path"], file)
        if os.path.exists(src):
            print(f"Copying {src} to {dst}")
            os.system(f"cat {src} >> {dst}")
        else:
            print(f"{src} does not exist")
    
```
