```python
import pandas as pd 
import os 

big_comparison = ["../Results/PRF/PRF_no_raster_NN_v2/results.csv",
                  "../Results/PRF/PRF_no_raster_v5/results.csv",
                  "../Results/PRF/PRF_dual_no_raster_NN_v3/results.csv",
                  "../Results/PRF/PRF_dual_no_raster_v5/results.csv"]

model = ["HybridGNet Unified", "HybridGNet Independent", "HybridGNet Dual Unified", "HybridGNet Dual Independent"]
adjacency = ["NonNaive", "Naive", "NonNaive", "Naive"]

df = pd.DataFrame()
for i, path in enumerate(big_comparison):
    df_temp = pd.read_csv(path)
    df_temp["Model"] = model[i]
    df_temp["Adjacency"] = adjacency[i]
    # replace values in organ column, Myo with Epi
    df_temp['organ'] = df_temp['organ'].replace({'LV Myo': 'LV Epi'})
    print(df_temp.columns)
    df = pd.concat([df,df_temp])

df.reset_index(drop=True,inplace=True)
```

```python
# Your original code that creates the 'results' table
def format_mean_std(group):
    return f"{group.mean():.3f} ± {group.std():.3f}"

df.organ = df.organ.str.title()  # Capitalize organ names

results = df.groupby(['organ', 'Model']).agg({
    'dc': format_mean_std,
    'hd': format_mean_std,
    'assd': format_mean_std
})
results.columns = ['DSC', 'HD', 'ASSD']

# Reshape: Organs as columns, Models as rows
reshaped = results.unstack(level=0)  # Move 'organ' from index to columns

# Reorder columns to group metrics by organ
organs = df.organ.unique()
metrics = ['DSC']

# Create new column order
new_columns = []
for organ in organs:
    for metric in metrics:
        new_columns.append((metric, organ))

# Reorder and display
reshaped_ordered = reshaped[new_columns]

print("=== Reshaped Table: Organs as Columns ===")
display(reshaped_ordered.T)

# latex
reshaped_ordered.T.to_latex("results.tex", float_format="%.2f", escape=False, multicolumn=True)
```

```python
aux = df.groupby(['Model']).agg({
    'dc': ['mean', 'std'],
    'hd': ['mean', 'std'],
    'assd': ['mean', 'std'],
}).reset_index().rename(columns={'dc': 'DSC', 'hd': 'HD', 'assd': 'ASSD'})
aux.to_latex("results_prf.tex", float_format="%.3f", escape=False, multicolumn=True, index=False)
```
